```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn import metrics

from warnings import simplefilter
%matplotlib inline
simplefilter(action='ignore', category=FutureWarning)

data = pd.read_csv("waterdataset.csv")
feature_cols = ['measurements','resp length','response memory count','response address','response memory','resp write fun','LL','H','sub function','command memory count']
X =data.loc[:, feature_cols].values


y = data['Target']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=10)

print("************STEP 1***************")
print("Traning Decision Tree Model")
print("                                 ")

#from sklearn.ensemble import RandomForestClassifier
#model1 = RandomForestClassifier(max_depth=2, random_state=0)
#from sklearn.neighbors import KNeighborsClassifier
#model1 = KNeighborsClassifier(n_neighbors=5)
model1 = DecisionTreeClassifier()
model1 = model1.fit(X_train,y_train)
y_pred= model1.predict(X_test)
print("Accuracy with test dataset",metrics.accuracy_score(y_test,y_pred))
results = confusion_matrix(y_test, y_pred) 
print ("Confusion Matrix :")
print(results) 

print("                                 ")
print("************STEP 2***************")
print("Traning Multi-layer perceptron Model")
print("                                 ")

y = data['result']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=10)

from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression(random_state=1)
#from sklearn.naive_bayes import MultinomialNB
#model2 = MultinomialNB()
#model2 = MLPClassifier(random_state=1, max_iter=300)
model2.fit(X_train, y_train)
y_predict = model2.predict(X_test)
print("Accuracy with test dataset: ",metrics.accuracy_score(y_test,y_predict))
results = confusion_matrix(y_test, y_predict) 
print ("Confusion Matrix :")
print(results) 


print("                                 ")
print("Working Model With new DATA ENTRY")
print("                                 ")

data = pd.read_csv("new.csv")
X =data.loc[[1], feature_cols].values
#0=Naïve Malicious Response Injection
#1=Complex Malicious Response Injection
#2=Malicious State Command Injection
#3=Malicious Parameter Command Injection
#4=Malicious Function Code Injection
#5=Attack Catagory: Denial of Service
#6=Attack Catagory: Reconnaissance 
y = data['Target']
y_predict = model1.predict(X)
if y_predict==1:
    print("*****IT IS AN ATTACK******")
    y = data['result']
    y_predict = model2.predict(X)
    if y_predict==1:
        print("Attack Catagory: Naïve Malicious Response Injection ")
    elif y_predict==2:
        print("Attack Catagory: Complex Malicious Response Injection ")
    elif y_predict==3:
        print("Attack Catagory: Malicious State Command Injection")
    elif y_predict==4:
        print("Attack Catagory:  Malicious Parameter Command Injection  ")
    elif y_predict==5:
        print("Attack Catagory: Malicious Function Code Injection")
    elif y_predict==6:
        print("Attack Catagory: Denial of Service  ")
    elif y_predict==7:
        print("Attack Catagory: Reconnaissance ")
    else:
        print("*****BUT CAN NOT BE DETERMINED*****")
else:
    print("******THIS DATA ENTRY IS NOT A ATTCAK******")
```
